## Mark all launch sites on a map


First, let's try to add each site's location on a map using site's latitude and longitude coordinates


The following dataset with the name `spacex_launch_geo.csv` is an augmented dataset with latitude and longitude added for each site. 


```python
#spacex_csv_file 
spacex_df_geo= pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv')
#spacex_df_geo=pd.read_csv(spacex_csv_file)
spacex_df_geo
```

Now, we can take a look at what are the coordinates for each site.


```python
# select relevant sub-columns
spacex_df_geo = spacex_df_geo[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df_geo.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

```python
#Initial the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)


# Empty dictionaries to store values if needed
lat = {}
long = {}
name = {}
coord = {}
circle = {}
marker = {}

# Loop over rows
for i, row in launch_sites_df.iterrows():
    # Access the correct values using column names
    lat[i] = row['Lat']
    long[i] = row['Long']
    name[i] = row['Launch Site']
    coord[i] = [lat[i], long[i]]

    # Add circle to the map
    circle[i] = folium.Circle(
        coord[i],
        radius=1000,
        color='#d35400',
        fill=True
    ).add_child(folium.Popup(name[i]))

    # Add marker with name
    marker[i] = folium.map.Marker(
        coord[i],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{name[i]}</b></div>'
        )
    )

    # Add both to the map
    site_map.add_child(circle[i])
    site_map.add_child(marker[i])

# Show the map after the loop
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
launch_site_lat = 28.562302	
launch_site_lon = -80.577356

coastline_lat = 28.56298
coastline_lon = -80.56794

coordinate_cos = [coastline_lat, coastline_lon]

distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
distance_coastline
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
distance_marker = folium.Marker(
    coordinate_cos,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
        )
    )
site_map.add_child(distance_marker)
```

```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
coordinates = [[launch_site_lat, launch_site_lon],coordinate_cos]
lines=folium.PolyLine(locations=coordinates, weight=1)
site_map.add_child(lines)
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site

#FOR RAILWAY AND KSC LC-39A
lat_site = 28.573255
long_site = -80.646895

railway_lat = 28.57398
railway_long = -80.65422

#calculating distance
railway_dist = calculate_distance(lat_site, long_site, railway_lat, railway_long)
railway_dist

#adding marker for the railway
railway_marker = folium.Marker(
    [railway_lat,railway_long],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(railway_dist),
        )
    )
site_map.add_child(railway_marker)

#adding the line
lines_railway = folium.PolyLine(locations = [[lat_site,long_site], [railway_lat, railway_long]], weight=1)
site_map.add_child(lines_railway)
```
